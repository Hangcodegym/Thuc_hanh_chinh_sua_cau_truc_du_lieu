# Đọc dữ liệu

```python
import pandas as pd
import numpy as np
df=pd.read_csv('FoodPrice_in_Turkey.csv', header=0, encoding='ISO-8859-1' )
display(df)
```

# Đổi tên thuộc tính

```python
df.rename(columns={'Place': 'Địa Điểm', 'ProductName':'Tên SP'}, inplace=True)
display(df)
```

# Thêm cột mới trong DATAFRAME

```python
# Thêm cột mới với tất cả giá trị rỗng:
df['New_column']='NaN'
df.head()
```

```python
# Thêm cột giảm giá 10% cho tất cả các bản ghi
df.insert(9, 'Giảm giá (%)', '10%' )
display(df)
```

```python
# Thêm cột giảm giá 12% cho tất cả các bản ghi
df.insert(10, 'Giảm giá 2 (%)', '12%' )
display(df)
```

# Thêm một dòng mới vào trong DataFrame

```python
df.loc[df.shape[0]]=['NA','RR','Rice',10,'KG',6, 2021, 84.3785, np.NaN, '10%','12%']
display(df)
```

# Xóa một cột trong dataframe

```python
del df['New_column']
display(df)
```

```python
# # Xóa một cột sử dụng phương thức pop()
# Phương thức pop trả về cột đã xóa
df1=df
print('df1 sau khi tạo')
display(df1)
df2=df1.pop('Giảm giá 2 (%)')
print('df1 sau khi xóa')
display(df1)
print('df2')
display(df2)
```

```python
# Xóa một cột sử dụng phương thức drop() 
df.drop('Giảm giá (%)', axis=1, inplace=True)
display(df)
```

```python
# Xóa nhiều cột sử dụng phương thức drop()
df.drop(['Month','Year'], axis=1, inplace=True)
display(df)
```
